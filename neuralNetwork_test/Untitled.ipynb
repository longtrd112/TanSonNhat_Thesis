```python
import folium
import json

# Define the corners of the rectangle
min_lat, min_lon = 10.8188 - 0.7, 106.652 - 0.7
max_lat, max_lon = 10.8188 + 0.7, 106.652 - 0.7
bbox = [[10.8188 - 0.7, 106.652 - 0.7], [10.8188 + 0.7, 106.652 + 0.7]]
popup_style = "font-family: Times New Roman, serif; font-size: 13px;"

# Create the map centered at the mean of the corner coordinates

m = folium.Map(location=[10.8188, 106.652], zoom_start=9.3,
               no_touch=True, zoom_control=False, scrollWheelZoom=False, dragging=False, doubleClickZoom=False)

folium.TileLayer('cartodbpositron', brightness=10, contrast=1).add_to(m)

with open('tsn_arrival.json', 'r') as f:
    data = json.load(f)
    
colors = ['blue', 'red', 'green', 'orange', 'purple', 'darkred', 'lightred', 'darkblue', 'lightblue', 'cadetblue', 'magenta',
         'darkviolet', 'hotpink', 'darkcyan', 'mistyrose', 'springgreen', 'orchid']

# Loop through each element in arrival_dict
for i, (key, value) in enumerate(data['arrival_dict'].items()):
    # Extract the list of waypoints
    waypoints = []
    for wp in value:
        waypoints.append(data['waypoint'][wp])
    # Create a PolyLine object with the list of waypoints and a different color
    folium.PolyLine(locations=waypoints, color=colors[i % len(colors)]).add_to(m)
    
folium.Circle(location=[10.8188, 106.652], radius=75000, color='black', fill_opacity=0.5).add_to(m)
icon_plane = folium.Icon(icon='plane', prefix='fa')

sw = [10.8188 - 0.75, 106.652 - 0.8]  
ne = [10.8188 + 0.75, 106.652 + 0.8]

folium.Rectangle(bounds=[sw, ne], color='black', fill_opacity=0.1, weight=0.5).add_to(m)

folium.Marker(location=[10.8188, 106.652], icon=icon_plane).add_to(m)

display(m)
```

```python
m.save('c.html')
```
