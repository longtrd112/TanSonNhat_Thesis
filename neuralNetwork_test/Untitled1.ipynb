```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as font_manager

# Define the Times New Roman font
font = font_manager.FontProperties(family='Times New Roman', size=13)

a = [178.1, 194]
b = [29.7, 42.4]

indices = np.arange(len(a))

# Set the width of the bars
width = 0.35
fig, ax = plt.subplots()

# Plot the first set of bars
ax.bar(indices - width/2, a, width=width, color='r', label='Mean Absolute Error')

# Plot the second set of bars
ax.bar(indices + width/2, b, width=width, color='b', label='Computation Time')

# Add labels and title with Times New Roman font
ax.set_ylabel('Seconds', fontproperties=font)
plt.xticks(indices, ['Multilayer Perceptrons', 'Long Short-Term Memory'], fontproperties=font)
plt.yticks(fontproperties=font)
for i, v in enumerate(a):
    ax.text(i - width/2, v + 3, str(v), color='black', ha='center', fontproperties=font)
for i, v in enumerate(b):
    ax.text(i + width/2, v + 3, str(v), color='black', ha='center', fontproperties=font)
# Set the legend font to Times New Roman
ax.legend()
fig.set_size_inches(10, 6)
ax.set_title('Model performance comparison - Third approach', fontproperties=font)
plt.tight_layout()
# Show the plot
plt.show()
```
